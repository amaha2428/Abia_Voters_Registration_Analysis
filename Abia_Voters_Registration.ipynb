### Data cleaning

```python
# removing first and last row of our dataset
df.drop([0,19], inplace=True)
```

```python
df
```

```python
df.drop(columns=['Unnamed: 1',1], axis=1, inplace=True)
```

```python
df
```

```python
# updating columns
df.columns = df.iloc[0]
```

```python
# removing row 1 since column has already been updated
df.drop([1], axis=0, inplace=True)
```

```python
# rearranging indexing of rows
df.reset_index(inplace=True)
```

```python
df
```

```python
df.columns
```

```python
# removing the index column created
df.drop('index', axis=1, inplace=True)
```

```python
# renaming columns
df.rename(columns={
    'No of Registered Voters' : 'no_reg_voters',
    'No of Collected PVCs' : 'no_coll_pvc',
    'No of Uncollected PVCs' : 'no_uncoll_pvc',
    'Percentage of Collected PVCs to Registered Voters' : 'per_coll_pvc'
}, inplace=True)
```

```python
# changing columns data types

df['no_reg_voters'] = df['no_reg_voters'].astype(int)
df['no_coll_pvc'] = df['no_coll_pvc'].astype(int)
df['no_uncoll_pvc'] = df['no_uncoll_pvc'].astype(int)
df['per_coll_pvc'] = df['per_coll_pvc'].astype(float)
```

```python
df['per_coll_pvc'] = df['per_coll_pvc'].apply(lambda x: x*100)
```

```python
df
```

```python
df.info()
```


## Visualization and Analysis

### Local Government Areas with most Registered Voters

```python
# function for horizontal bar plot
def barh_plot(column1, column2, title, xlabel, ylabel):
    # sorting the plots in descending order
    sorted_data = sorted(zip(df[column1], df[column2]), key=lambda x: x[1], reverse=True)
    categories = [x[0] for x in sorted_data]
    values = [x[1] for x in sorted_data]
    
    #
    plt.figure(figsize=(10,8))
    plt.barh(categories, values)
    plt.title(title)
    plt.ylabel(ylabel)
    plt.xlabel(xlabel)
    plt.gca().invert_yaxis()
```

```python
barh_plot('LGA', 'no_reg_voters',
         'Local Government Area with most Registered voters',
         'No of Registered voters', 'LGA')
```

### Local Government Area with number of uncollected PVC

```python
barh_plot('LGA', 'no_uncoll_pvc',
         "LGA with the most uncollected PVC's",
         "count of uncollected PVC's", 'LGA')
```

### LGA with most collected PVC's

```python
barh_plot('LGA', 'no_coll_pvc',
        "LGA with the most collected PVC's",
        "count of uncollected PVC's", "LGA")
```

### LGA with percentage of collected PVC's

```python
barh_plot('LGA', 'per_coll_pvc',
         "LGA with percentage of collected PVC's",
         "percentage of PVC's", "LGA")
```

### Observations
- ABA SOUTH LGA had more registered voters of about 250,000 people for the 2023 Presidential election followed by ABA NORTH.
- Despite being the region with most registered individuals, ABA SOUTH is also the region with low turn out for PVC collection.
- ABA NORTH was the second region with low turn out for PVC collection but is still the region with most collected PVC's.
- When viewed by percentage, AROCHUWU appeared to be the region with most collected PVC's with 99.9% of PVC's collected.
